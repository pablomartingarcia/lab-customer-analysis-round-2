5. Check and deal with NaN values.

```python
data1 = df.copy()
```

```python
data1.isnull()
```

```python
data1.isnull().sum()
```

```python
data1['STATE'] = data1['STATE'].fillna('California')
```

```python
data1.isnull().sum()
```

```python
value_counts = df['STATE'].value_counts()
print(value_counts)
```

```python
value_counts = df['RESPONSE'].value_counts()
print(value_counts)
```

```python
data1['RESPONSE'] = data1['RESPONSE'].fillna('No')
```

```python
value_counts = df['MONTHS_SINCE_LAST_CLAIM'].value_counts()
print(value_counts)
```

```python
data1['MONTHS_SINCE_LAST_CLAIM'] = data1['MONTHS_SINCE_LAST_CLAIM'].fillna(data1['MONTHS_SINCE_LAST_CLAIM'].mean())
```

```python
data1['NUMBER_OF_OPEN_COMPLAINTS'] = data1['NUMBER_OF_OPEN_COMPLAINTS'].fillna(data1['NUMBER_OF_OPEN_COMPLAINTS'].mean())
```

```python
value_counts = df['VEHICLE_CLASS'].value_counts()
print(value_counts)
```

```python
data1['VEHICLE_CLASS'] = data1['VEHICLE_CLASS'].fillna('Four-Door Car')
```

```python
value_counts = df['VEHICLE_SIZE'].value_counts()
print(value_counts)
```

```python
data1['VEHICLE_SIZE'] = data1['VEHICLE_SIZE'].fillna('Medsize')
```

```python
value_counts = df['VEHICLE_TYPE'].value_counts()
print(value_counts)
```

```python
data1['VEHICLE_TYPE'] = data1['VEHICLE_TYPE'].fillna('A')
```
